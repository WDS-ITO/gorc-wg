# Types of commons

#### This notebook holds the converted data for representing types of commons that have participated in the GORC-WG speaker series or otherwise investigated in support of creating GORC-WG outputs.
#### Find the reasoning behind this chart and explanations for the visual representation in the GORC-WG martyr doc: https://docs.google.com/document/d/1c34WsuG2tm_dFIWhqgwHKM8T3L1ws2t4cceClRxWrO0/edit?usp=sharing

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
```

```python
# horizontal axis = maturity (young =1)
# vertical axis = weight class (lightweight = 1)
# colours = reach (red=domain, green = national, blue = pan-national --> orange=domain+national, purple=domain+pannational)
# shapes = service level (square=dispersed, triangle=collected, circle=centralised)

legend_elements=[Line2D([0],[0],c='white', markerfacecolor='black',markersize=10,marker='s',label='Dispersed Services'),
                 Line2D([0],[0],c='white', markerfacecolor='black',markersize=10, marker='^',label='Collected Services'),
                 Line2D([0],[0],c='white', markerfacecolor='black',markersize=10,marker='o',label='Centralized Services'),
                Line2D([0],[0],c='blue',lw=4,label='Pan-national'),
                 Line2D([0],[0],c='purple',lw=4,label='Domain-specific and Pan-national'),
                 Line2D([0],[0],c='red',lw=4,label='Domain-specific'),
                 Line2D([0],[0],c='orange',lw=4,label='Domain-specific and National'),
                 Line2D([0],[0],c='green',lw=4,label='National')]

fig=plt.figure()
ax=fig.add_subplot(111)
plt.scatter([3], [1], marker='s', c='purple', s=200, label="IVOA")
ax.text(2.925,1.1,'IVOA',None)
plt.scatter([1], [2], marker='o', c='green', s=200, label='NII RCOS')
ax.text(0.925,1.8,'NII RCOS',None)
plt.scatter([2], [3], marker='^', c='blue', s=200, label="EOSC")
ax.text(1.925,2.8,'EOSC',None)
plt.scatter([2], [2], marker='^', c='green',s=200, label="ARDC")
ax.text(1.925,2.1,'ARDC',None)
plt.scatter([1.1], [2], marker='^', c='green',s=200, label="KISTI KRDC")
ax.text(1.025,2.1,'KISTI KRDC',None)
plt.scatter([1], [1], marker='^', c='orange',s=200, label="CCADI")
ax.text(0.925,1.1,'CCADI',None)
plt.xticks([1,2,3])
plt.yticks([1,2,3])
ax.set_ylabel('Weight Class')
ax.set_xlabel('Maturity')
lgd=ax.legend(handles=legend_elements, bbox_to_anchor=(1,.8))

#plt.show()
plt.savefig('TypesofCommons.png', bbox_inches='tight')
```

```python
import plotly.graph_objects as go
import numpy as np

x=np.array([3,1,2,2,1,1]) #Maturity
y=np.array([1,2,3,2,2,1]) #weight class
z=np.array([1,3,2,2,2,2]) # service level
c=np.array(['purple','green','blue','green','green','orange']) #reach
commons=np.array(['IVOA', 'NII RCOS', 'EOSC','ARDC','KISTI KRDC','CCADI'])

fig = go.Figure(data=[go.Scatter3d(x=x, y=y, z=z, text=commons, 
                                   hovertemplate = '<b>%{text}</b><br><br>'+
                                   'Maturity: %{x}<br>'+'Weight Class: %{y}<br>'+'Service Level: %{z}'+'<extra></extra>',
                                   showlegend=False,
                                   mode='markers',
                                  marker=dict(
        size=10,
        color=c,                # set color to an array/list of desired values
        opacity=0.8
    ))])

fig.update_layout(title={
    'text':'Commons reviewed by GORC-WG',
    'y':0.9,
    'x':0.35,
    'xanchor':'center',
    'yanchor':'top'})

fig.update_layout(scene = dict(
                    xaxis_title='Maturity',
                    yaxis_title='Weight Class',
                    zaxis_title='Service Level'),
                    width=800,
                  height=800,
                    margin=dict(r=20, b=10, l=20, t=20))

fig.update_layout(scene=dict( xaxis = dict(
                        ticktext= ['Young','Middle Aged','Mature'],
                        tickvals= [1,2,3],
                        tickangle=-90,
                        range=[0.75,3.25]),
                    yaxis = dict(
                        ticktext= ['Lightweight','Middleweight','Heavyweight'],
                        tickvals= [1,2,3],
                        tickangle=-90,
                        range=[0.75,3.25]),
                    zaxis = dict(
                        ticktext= ['Dispersed','Collected','Centralized'],
                        tickvals= [1,2,3],
                        tickangle=-90,
                        range=[0.75,3.25])))

fig.add_trace(go.Scatter3d(
    x=[0],
    y=[0],
    z=[0],
    mode='markers',
    marker=dict(color='blue'),
    name="Pan-National"
))

fig.add_trace(go.Scatter3d(
    x=[0],
    y=[0],
    z=[0],
    mode='markers',
    marker=dict(color='purple'),
    name="Domain-specific & Pan-National"
))

fig.add_trace(go.Scatter3d(
    x=[0],
    y=[0],
    z=[0],
    mode='markers',
    marker=dict(color='red'),
    name="Domain-specific"
))

fig.add_trace(go.Scatter3d(
    x=[0],
    y=[0],
    z=[0],
    mode='markers',
    marker=dict(color='orange'),
    name="Domain-specific & National"
))

fig.add_trace(go.Scatter3d(
    x=[0],
    y=[0],
    z=[0],
    mode='markers',
    marker=dict(color='green'),
    name="National"
))

fig.write_html("index.html")
#fig.show()
```
